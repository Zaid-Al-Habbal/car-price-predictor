# Get the vehicles dataset

```python
import kagglehub
import pandas as pd
import os
```

```python
data_path = kagglehub.dataset_download("nehalbirla/vehicle-dataset-from-cardekho")
```

```python
csv_path = os.path.join(data_path, "Car details v3.csv")
```

```python
df = pd.read_csv(csv_path)
df.head()
```

```python
from pathlib import Path


to_path = Path("../datasets/cars.csv")
to_path.parent.mkdir(parents=True, exist_ok=True)

df.to_csv(to_path, index=False)
```
